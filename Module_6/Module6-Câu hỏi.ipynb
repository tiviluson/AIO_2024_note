## Mô tả bài toán

Cho một mạng có kiến trúc như hình vẽ trên, gồm 2 nhánh:
   - **Nhánh chính**: Bao gồm 3 lớp convolution, mỗi lớp đều có BatchNorm và ReLU theo sau.
   - **Nhánh Skip Connection:** Nhánh này sử dụng một lớp convolution (có BatchNorm và ReLU theo sau) để giảm kích thước không gian của tensor đầu vào phù hợp với kích thước của tensor đầu ra từ nhánh chính.

**Lưu ý**: Các lớp convolution trong nhánh chính nếu không được nêu rõ trong hình, thì mặc định stride = 1 và padding = 0.
