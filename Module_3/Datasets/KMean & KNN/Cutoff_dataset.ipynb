```python
df = pd.read_csv('/content/drive/MyDrive/Exam Module 3/Datasets/KMean & KNN/data.csv')
df.head()
```

```python
def min_max_normalize(column):
    return (column - column.min()) / (column.max() - column.min())

df['perimeter_mean'] = min_max_normalize(df['perimeter_mean']).round(2)
df['area_mean'] = min_max_normalize(df['area_mean']).round(2)
df['compactness_mean'] = min_max_normalize(df['compactness_mean']).round(2)
```

```python
features_selected = ["perimeter_mean", "area_mean", "compactness_mean", "diagnosis"]
data = df[features_selected]
data
```

```python
b_samples = data[data['diagnosis'] == 'B'].sample(n=5, random_state=42)
m_samples = data[data['diagnosis'] == 'M'].sample(n=5, random_state=42)

selected_samples = pd.concat([b_samples, m_samples])
selected_samples = selected_samples.reset_index(drop=True)
selected_samples
```

```python
test_data = data.sample(n=5, random_state=42)
test_data = test_data.reset_index(drop=True)
test_data
```

```python
selected_samples.to_csv("/content/drive/MyDrive/Exam Module 3/Datasets/KMean & KNN/final_dataset.csv", index=False)
```
