## Câu hỏi tính toán 1


```python
import torch
import torch.nn as nn

class Flow(nn.Module):
    def __init__(self) -> None:
        super().__init__()

        self.linear = nn.Linear(2, 1)

        with torch.no_grad():
            self.linear.weight.copy_(torch.tensor([[0.7, 0.4]]))
            self.linear.bias.copy_(torch.tensor([0]))

            self.linear.weight.requires_grad = False
            self.linear.bias.requires_grad = False

    def forward(self,  t: torch.Tensor, x: torch.Tensor) -> torch.Tensor:
        t = t.unsqueeze(0).float()
        x = torch.cat([x, t], dim=0)
        x = self.linear(x)
        x = nn.Tanh()(x)
        return x
```

```python
# Khởi tạo model - loss
model = Flow()
loss_fn = nn.MSELoss()

# Giá trị cho trước
x_1 = torch.tensor([0.5])
x_0 = torch.tensor([-0.3])
t = torch.tensor(0.6)
###

# Tính toán
x_t = (1 - t) * x_0 + t*x_1
print("x_t:", x_t.item())

pred = model(t, x_t)
print("\npred:", pred.item())

target = x_1 - x_0
print("\ntarget:", target.item())

# Tính toán loss
loss = loss_fn(pred, target)
print("\n\n\n***Đáp án***\nloss:  ", loss.item())
```

## Câu hỏi tính toán 2


```python
class Flow2(nn.Module):
    def __init__(self) -> None:
        super().__init__()

        self.linear = nn.Linear(2, 1)

        with torch.no_grad():
            self.linear.weight.copy_(torch.tensor([[0.2, 0.1]]))
            self.linear.bias.copy_(torch.tensor([0]))

            self.linear.weight.requires_grad = False
            self.linear.bias.requires_grad = False

    def forward(self,  t: torch.Tensor, x: torch.Tensor) -> torch.Tensor:
        t = t.unsqueeze(0).float()
        x = torch.cat([x, t], dim=0)
        x = self.linear(x)
        x = nn.Tanh()(x)
        return x
```

```python
# Khởi tạo model
model = Flow2()

# Các giá trị cho trước
x1 = torch.tensor([0.3])
steps = 10
###

delta = 1/steps
t1 = torch.tensor(0) + delta
print("t1:", t1)

# Tính toán x2
x2 = x1 + delta * model(t1, x1)
print("x2:", x2.item())
t2 = t1 + delta
print("t2:", t2)

# Tính toán x3
x3 = x2 + delta * model(t2, x2)
print("\n\n\n***Đáp án***\nx3:  ", x3.item())
```
